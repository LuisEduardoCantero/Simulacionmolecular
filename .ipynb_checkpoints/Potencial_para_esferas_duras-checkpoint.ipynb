```python
def potencialwca(x1, lamb, sigma, t):
    d = []
    i=0
    a = lamb * np.power(lamb / (lamb - 1), lamb - 1)
    b = lamb / (lamb - 1)
    while i <= (len(x1) - 1):
        j = i + 1
        while j < len(x1):
            r = x1[i] - x1[j]            
            if r < (sigma * b):
                pot = (a * (np.power((sigma / r), lamb) - np.power((sigma / r), (lamb - 1))) + 1)
            if r >= (sigma * b):
                pot = 0            
            d.append(pot)
            j = j + 1
        i= i + 1
    return np.sum(d)
```

```python
b  = potencialwca(x, 50, 1.0, 1.44)
b1  = potencialwca(x1, 50, 1.0, 1.44)
b2  = potencialwca(x2, 50, 1.0, 1.44)
b3  = potencialwca(x3, 50, 1.0, 1.44)
b4  = potencialwca(x4, 50, 1.0, 1.44)

b0 = [b, b1, b2, b3, b4]
```
