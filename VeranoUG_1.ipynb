# Potenciales Hard Sphere Double Yukawa (HSDY) y Leonard-Jones-Yukawa) LJY

```python
# Importamos librerias
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
hsdy1 = np.array([10, 0.5, 0.1, 1.2424])
x1 = np.arange(1.0,hsdy1[3],0.01)

hsdy2 = np.array([11.3023, 0.5, 0.094515, 1.2184])
x2 = np.arange(1.0,hsdy2[3],0.01)

hsdy3 = np.array([11.9279, 0.5, 0.092307, 1.2085])
x3 = np.arange(1.0,hsdy3[3],0.01)

hsdy4 = np.array([8.134, 0.5, 0.2, 1.2108])
x4 = np.arange(1.0,hsdy4[3],0.01)

hsdy5 = np.array([20, 0.5, 0.175, 1.0894])
x5 = np.arange(1.0,hsdy5[3],0.01)

hsdy6 = np.array([5, 0.5, 0.2, 1.3577])
x6 = np.arange(1.0,hsdy6[3],0.01)

hsdy7 = np.array([10, 0.5, 0.01, 1.4847])
x7 = np.arange(1.0,hsdy7[3],0.01)

hsdy8 = np.array([10, 0.5, 0.05, 1.3153])
x8 = np.arange(1.0,hsdy8[3],0.01)
```

```python
T = 1
y1 = (1/(T * (1 - hsdy1[2]) * x1))*(- np.exp( -hsdy1[0] * (x1 - 1)) + hsdy1[2] * np.exp(-hsdy1[1] * (x1 - 1)))
y2 = (1/(T * (1 - hsdy2[2]) * x2))*(- np.exp( -hsdy2[0] * (x2 - 1)) + hsdy2[2] * np.exp(-hsdy2[1] * (x2 - 1)))
y3 = (1/(T * (1 - hsdy3[2]) * x3))*(- np.exp( -hsdy3[0] * (x3 - 1)) + hsdy3[2] * np.exp(-hsdy3[1] * (x3 - 1)))
y4 = (1/(T * (1 - hsdy4[2]) * x4))*(- np.exp( -hsdy4[0] * (x4 - 1)) + hsdy4[2] * np.exp(-hsdy4[1] * (x4 - 1)))
y5 = (1/(T * (1 - hsdy5[2]) * x5))*(- np.exp( -hsdy5[0] * (x5 - 1)) + hsdy5[2] * np.exp(-hsdy5[1] * (x5 - 1)))
y6 = (1/(T * (1 - hsdy6[2]) * x6))*(- np.exp( -hsdy6[0] * (x6 - 1)) + hsdy6[2] * np.exp(-hsdy6[1] * (x6 - 1)))
y7 = (1/(T * (1 - hsdy7[2]) * x7))*(- np.exp( -hsdy7[0] * (x7 - 1)) + hsdy7[2] * np.exp(-hsdy7[1] * (x7 - 1)))
y8 = (1/(T * (1 - hsdy8[2]) * x8))*(- np.exp( -hsdy8[0] * (x8 - 1)) + hsdy8[2] * np.exp(-hsdy8[1] * (x8 - 1)))
```

```python
plt.plot(x1,y1, "-", color = 'g', label="HSDY1")
plt.plot(x2,y2, "-", color = 'm', label="HSDY2")
plt.plot(x3,y3, "-", color = 'b', label="HSDY3")
plt.plot(x4,y4, "-", color = 'r', label="HSDY4")
plt.plot(x5,y5, "-", color = 'c', label="HSDY5")
plt.plot(x6,y6, "-", color = 'y', label="HSDY6")
plt.plot(x7,y7, "-", color = 'b', label="HSDY7")
plt.plot(x8,y8, "-", color = 'm', label="HSDY8")
plt.legend()
```

```python
ljy1 = np.array([16.6755, 0.5, 0.090982, 1.2665])
r1 = np.arange(1.0, ljy1[3], 0.001)

ljy2 = np.array([32, 0.5, 0.17, 1.1005])
r2 = np.arange(1.0 ,ljy2[3], 0.001)

ljy3 = np.array([100, 0.5, 0.2, 1.0282])
r3 = np.arange(1.0, ljy3[3], 0.001)

ljy4 = np.array([100, 0.5, 0.05, 1.0433])
r4 = np.arange(1.0, ljy4[3], 0.001)

ljy5 = np.array([18, 2.0, 8.0, 1.1358])
r5 = np.arange(1.0, ljy5[3], 0.001)
```

```python
u1 = (1/T) * (4 * (r1 ** (-2 * ljy1[0]) - r1 ** (-ljy1[0])) + ljy1[2] * (1/ljy1[1]) / r1 * np.exp(-r1 * ljy1[1]))
u2 = (1/T) * (4 * (r2 ** (-2 * ljy2[0]) - r2 ** (-ljy2[0])) + ljy2[2] * (1/ljy2[1]) / r2 * np.exp(-r2 * ljy2[1]))
u3 = (1/T) * (4 * (r3 ** (-2 * ljy3[0]) - r3 ** (-ljy3[0])) + ljy3[2] * (1/ljy3[1]) / r3 * np.exp(-r3 * ljy3[1]))
u4 = (1/T) * (4 * (r4 ** (-2 * ljy4[0]) - r4 ** (-ljy4[0])) + ljy4[2] * (1/ljy4[1]) / r4 * np.exp(-r4 * ljy4[1]))
u5 = (1/T) * (4 * (r5 ** (-2 * ljy5[0]) - r5 ** (-ljy5[0])) + ljy5[2] * (1/ljy5[1]) / r5 * np.exp(-r5 * ljy5[1]))
```

```python
plt.plot(r1, u1, "-", color = 'm', label="LJY1")
plt.plot(r2, u2, "-", color = 'b', label="LJY2")
plt.plot(r3, u3, "-", color = 'g', label="LJY3")
plt.plot(r4, u4, "-", color = 'y', label="LJY4")
plt.plot(r5, u5, "-", color = 'c', label="LJY5")
plt.legend()
plt.rcParams["figure.figsize"] = (15, 20)
```
