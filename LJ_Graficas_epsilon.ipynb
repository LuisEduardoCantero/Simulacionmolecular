# Computer simulation of liquids

```python
# Importamos librerias
import numpy as np
import matplotlib.pyplot as plt
```

### Potencial Lennard-Jones

 $u(r) = 4ε * \left[ \left(\frac{1}{x} \right)^{12} - \left(\frac{1}{x} \right)^{6} \right] $ 

donde:

$x = r/σ$

```python
x = np.arange(0.01,5,0.01)
eps1 = 0.1
eps2 = 0.5
eps3 = 1.0
eps4 = 2.0
eps5 = 5.0
eps6 = 10.0

y1 = 4 * eps1 * ((1/x) ** 12 - (1/x) ** 6)
y2 = 4 * eps2 * ((1/x) ** 12 - (1/x) ** 6)
y3 = 4 * eps3 * ((1/x) ** 12 - (1/x) ** 6)
y4 = 4 * eps4 * ((1/x) ** 12 - (1/x) ** 6)
y5 = 4 * eps5 * ((1/x) ** 12 - (1/x) ** 6)
y6 = 4 * eps6 * ((1/x) ** 12 - (1/x) ** 6)
```

```python
plt.plot(x,y1, "-", color = 'g', label="ε = 0.1")
plt.plot(x,y2, "-", color = 'r', label="ε = 0.5")
plt.plot(x,y3, "-", color = 'b', label="ε = 1.0")
plt.plot(x,y4, "-", color = 'c', label="ε = 2.0")
plt.plot(x,y5, "-", color = 'y', label="ε = 5.0")
plt.plot(x,y6, "-", color = 'm', label="ε = 10.0")

plt.xlabel("r/σ")
plt.ylabel("β U(V)")
plt.ylim(-11, 6)
plt.grid(b=None, which='major', axis='both')
plt.rcParams["figure.figsize"] = (15,10)
plt.legend()
```

### Fuerza

Podemos reescribir el potencial LJ de la sigueinte forma:

 $u(r) = 4ε * \left[ \left(\frac{r}{σ} \right)^{-12} - \left(\frac{r}{σ} \right)^{-6} \right] $ 

El campo de fuerza resultante de este potencial esta dado por:

 $F(r) = -\frac{∂u}{∂r} = \frac{48ε}{σ} * \left[ \left(\frac{r}{σ} \right)^{-13} - \frac{1}{2} \left(\frac{r}{σ} \right)^{-7} \right] $ 

 $F(r) = -\frac{∂u}{∂r} = 48ε * \left[ \left(x \right)^{-13} - \frac{1}{2} \left(x \right)^{-7} \right] $ 

Donde:

$𝑥=𝑟/σ$

```python
y7 = 48 * eps1 * ((x) ** -13 - 0.5 * (x) ** -7)
y8 = 48 * eps2 * ((x) ** -13 - 0.5 * (x) ** -7)
y9 = 48 * eps3 * ((x) ** -13 - 0.5 * (x) ** -7)
y10 = 48 * eps4 * ((x) ** -13 - 0.5 * (x) ** -7)
y11 = 48 * eps5 * ((x) ** -13 - 0.5 * (x) ** -7)
y12 = 48 * eps6 * ((x) ** -13 - 0.5 * (x) ** -7)
```

```python
plt.plot(x,y1, "-", color = 'g', label="Energia potencial, ε = 0.1")
plt.plot(x,y7, "-", color = 'r', label="Fuerza, ε = 0.1")
plt.ylim(-.5, 1)
plt.xlabel("r/σ")
plt.ylabel("β U(V), F(r)")
plt.grid(b=None, which='major', axis='both')
plt.rcParams["figure.figsize"] = (15,10)
plt.legend()
```
