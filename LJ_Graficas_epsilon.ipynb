```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0.01,5,0.01)
eps1 = 0.1
eps2 = 0.5
eps3 = 1.0
eps4 = 2.0
eps5 = 5.0
eps6 = 10.0

y1 = 4 * eps1 * ((1/x) ** 12 - (1/x) ** 6)
y2 = 4 * eps2 * ((1/x) ** 12 - (1/x) ** 6)
y3 = 4 * eps3 * ((1/x) ** 12 - (1/x) ** 6)
y4 = 4 * eps4 * ((1/x) ** 12 - (1/x) ** 6)
y5 = 4 * eps5 * ((1/x) ** 12 - (1/x) ** 6)
y6 = 4 * eps6 * ((1/x) ** 12 - (1/x) ** 6)
```

```python
plt.plot(x,y1, "-", color = 'g', label="ε = 0.1")
plt.plot(x,y2, "-", color = 'r', label="ε = 0.5")
plt.plot(x,y3, "-", color = 'b', label="ε = 1.0")
plt.plot(x,y4, "-", color = 'c', label="ε = 2.0")
plt.plot(x,y5, "-", color = 'y', label="ε = 5.0")
plt.plot(x,y6, "-", color = 'm', label="ε = 10.0")

plt.xlabel("r/σ")
plt.ylabel("β U(V)")
plt.ylim(-11, 6)
plt.grid(b=None, which='major', axis='both')
plt.rcParams["figure.figsize"] = (15,10)
plt.legend()
```
